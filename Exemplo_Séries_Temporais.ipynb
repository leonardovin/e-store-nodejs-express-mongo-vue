```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("petro.csv", delimiter=',', index_col='Date')

df
```

```python
print(df.shape)
```

```python
step_forward = 2 #etapas a frente

if df.iloc[0:1, 3].values[0] != 0:
  df['Close'] = df.Close.shift(periods=step_forward, fill_value=0)  #Só pode executar uma vez.

print(df.head())
print(df.shape)
```

```python
df2 = df.iloc[step_forward:, :]

df2
```


Essa configuração considera N dias de treino e 1 dia de teste para prever ETAPAS DE TEMPO À FRENTE.

Por exemplo: Suponha que todo conjunto do dataset é 02/01/2021 à 30/01/2021

1ª Avaliação

    TREINO: 02 à 11 de Janeiro
    TESTE: 12 de Janeiro. 
    Para prever ETAPA DE TEMPO À FRENTE. 

Se step-forward = 1 
A previsão é para 13 de Janeiro.

Se step-forward = 2 
A previsão é para 14 de Janeiro.


2ª Avaliação

    TREINO: 02 à 12 de Janeiro
    TESTE: 13 de Janeiro. 
    Para prever ETAPA DE TEMPO À FRENTE. 

Se step-forward = 1 
A previsão é para 14 de Janeiro.

Se step-forward = 2 
A previsão é para 15 de Janeiro.

...

**n** Avalições


```python


train = int(df.shape[0] * 0.3) # Primeiro treino com 30% dos dados

print("1º Treino: 0 - ", train)

for day in range(train, df.shape[0]):

  X_treino = df2.iloc[0: day, [0, 1, 2, 4, 5]]
  y_treino = df2.iloc[0: day, 4]

  X_teste = df2.iloc[day : day + 1, [0, 1, 2, 4, 5] ]
  y_teste = df2.iloc[day : day + 1, 4]

  print("Dados de treino X: \n", X_treino)
  print("Dados de treino y: \n", y_treino)
  print("Dados de teste X: \n", X_teste)
  print("Dados de teste y: \n", y_teste)

  #APENAS PARA NAO EXECUTAR TUDO
  if(day == 806):
    break

```
